```python
#json 변환(단어: tfidf )
def display_scores(vectorizer, tfidf_result, d_num):
    # http://stackoverflow.com/questions/16078015/
    scores = zip(vectorizer.get_feature_names(),
                 np.asarray(tfidf_result.sum(axis=0)/d_num).ravel())
    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)
    return sorted_scores
```

```python
for i in category_dict.keys():
    vectorizer = TfidfVectorizer()
    x_train = vectorizer.fit_transform(json_corpus[i])
    print(len(vectorizer.vocabulary_))
    tmp = display_scores(vectorizer, x_train,len(json_corpus[i]))
    name = './'+str(category_dict[i])+'_voca.json'

    with open(name,'w') as f:
        json.dump(dict(tmp),f)
```
