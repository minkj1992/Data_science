```python
from sklearn.linear_model import LogisticRegression, SGDClassifier, Perceptron
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.utils import shuffle
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle
import json
import numpy as np
import random

category_dict = {"안전/환경": 1,"미래": 2,"일자리": 3,"보건복지": 4,"정치개혁": 5,"경제민주화": 6,"인권/성평등": 7,"외교/통일/국방": 8,"육아/교육": 9,"문화/예술/체육/언론": 10,"반려동물": 11,"교통/건축/국토": 12,"행정": 13,"농산어촌": 14,"저출산/고령화대책": 15,"성장동력": 16,"기타": 17}
# tf_idf load
#vectorizing 한 pickle 파일을 불러온다.
# 육아/교육": 9,"문화/예술/체육/언론": 10,"반려동물": 11,"교통/건축/국토": 12,"행정": 13,"농산어촌": 14,"저출산/고령화대책": 15,"성장동력": 16,"기타": 17}
##here
category = "성장동력"


# 토큰화 된 corpus 데이터를 불러온다.
with open('corpus2.json','r') as f:
    json_corpus = json.load(f)

a_list = list()
b_list = list()
c_list = list()
d_list = list()

for i in category_dict.keys():
    if (i == "기타"):
        break
        
    print(str(i)+" is in train case")
    vectorizer = TfidfVectorizer()
    
    data = json_corpus[i]
    vectorizer.fit(data)
    y_train = [1]*len(data)
#     100개만
    x_test = json_corpus[i][:100]
    y_test = [1]*100
    
    data.extend(json_corpus['기타'])
#     x_train = vectorizer.fit_transform(data)
    x_train = vectorizer.transform(data)
    y_train.extend([0]*len(json_corpus['기타']))
    
    
    #랜덤 카테고리 뽑아내기
    category_list = list(category_dict.keys())
    category_list.remove("기타")
    category_list.remove(i)
    test_cate = random.sample(category_list,4)
    print(str(test_cate)+" is in test case")
    
    for cat in test_cate:
        x_test.extend(json_corpus[cat][:25])
#     print(x_test[-1])

    x_test = vectorizer.transform(x_test)
    y_test.extend([0]*100)

    X,Y = shuffle(x_train,y_train)
    x_test,y_test = shuffle(x_test,y_test)
    

    
    #여기서 x_test,y_test저장시켜주어야한다.

    print("#############################LogisticRegression##################################")
    # classifier LogisticRegression
    

    #[[해당카테고리, predict 맞았는지 틀렸는지]*200개씩] *16
    clf = LogisticRegression().fit(X, Y)
    sub = []
    print("훈련 세트 점수: {:.3f}".format(clf.score(X,Y)))
    
    for j,sentence in enumerate(x_test):
        prediction = clf.predict(sentence)
        k = str(prediction).strip('[')
        k = k.strip(']')
        sub.append([category_dict[i],int(int(k)==y_test[j])])
    a_list.append(sub)
    
    print("#############################Perceptron##########################################")
    # classifier Perceptron

    clf = Perceptron(random_state=0, tol=1e-5)
    clf.fit(X,Y)
    sub = []
    print("훈련 세트 점수: {:.3f}".format(clf.score(X,Y)))
    
    for j,sentence in enumerate(x_test):
        prediction = clf.predict(sentence)
        k = str(prediction).strip('[')
        k = k.strip(']')
        sub.append([category_dict[i],int(int(k)==y_test[j])])
        
    b_list.append(sub)

    print("#############################LinearSVC############################################")
    # classifier LinearSVC

    clf = LinearSVC(random_state=0, tol=1e-5)
    clf.fit(X,Y)
    sub = []
    print("훈련 세트 점수: {:.3f}".format(clf.score(X,Y)))
    
    for j,sentence in enumerate(x_test):
        prediction = clf.predict(sentence)
        k = str(prediction).strip('[')
        k = k.strip(']')
        sub.append([category_dict[i],int(int(k)==y_test[j])])
        
    c_list.append(sub)

    print("#############################SGDClassifier########################################")
    # classifier SGDClassifier
    # tol == early stopping

    clf = SGDClassifier(max_iter=1000, tol=1e-5)
    clf.fit(X,Y)
    sub = []
    print("훈련 세트 점수: {:.3f}".format(clf.score(X,Y)))
    
    for j,sentence in enumerate(x_test):
        prediction = clf.predict(sentence)
        k = str(prediction).strip('[')
        k = k.strip(']')
        sub.append([category_dict[i],int(int(k)==y_test[j])])
        
    d_list.append(sub)

    print(str(i)+"category train is finished!")
    print("############################################################################")
    print("############################################################################")
    print("############################################################################")
   
```

```python
name = './LogisticRegression.json'
with open(name,'w') as f:
    json.dump(a_list,f)
```

```python
name = './Perceptron.json'
with open(name,'w') as f:
    json.dump(b_list,f)
```

```python
name = './LinearSVC.json'
with open(name,'w') as f:
    json.dump(c_list,f)
```

```python
name = './SGDClassifier.json'
with open(name,'w') as f:
    json.dump(d_list,f)
```
